```python
import pandas as pd
```

```python
df_vac = pd.read_csv('vaccinations.csv')

df_ep = pd.read_csv('epidemiology.csv')

df_index = pd.read_csv('index.csv')

df_mob = pd.read_csv('mobility.csv')
```

```python
df_mob.head()
```

```python
# drop mobility_retail_and_recreation column and convert date to datetime
df_mob = df_mob.drop(columns=['mobility_retail_and_recreation'])
df_mob.head()
```

```python
df_vac.head()
```

```python
df_vac = df_vac[['location_key', 'date', 'new_persons_fully_vaccinated']]
```

```python
df_ep.head()
```

```python
df_ep = df_ep[['location_key', 'date', 'new_confirmed', 'new_deceased', 'new_recovered']]
```

```python
df_index.columns
```

```python
df_index = df_index[
    ["location_key", "iso_3166_1_alpha_2", "country_name"]
]
df_index.value_counts()
```

```python
df_combined = pd.merge(df_index, df_vac, on=["location_key"], how="left")

df_combined = pd.merge(df_combined, df_ep, on=["location_key", "date"], how="left")

df_combined = pd.merge(df_combined, df_mob, on=["location_key", "date"], how="left")

df_combined.drop(df_combined[df_combined['location_key'] != df_combined['iso_3166_1_alpha_2']].index, inplace=True)
```

```python
df_combined
```

```python
df_combined = df_combined[~df_combined.date.isna()]
df_combined = df_combined[~df_combined.country_name.isna()]
```

```python
df_mobility = df_combined[['date', 'country_name', 'new_confirmed', 'new_persons_fully_vaccinated', 'new_deceased', 'new_recovered', 'mobility_grocery_and_pharmacy', 'mobility_parks', 'mobility_transit_stations', 'mobility_workplaces', 'mobility_residential']]
```

```python
df_mobility.fillna(0, inplace=True)
```

```python
df_mobility.to_csv('mob.csv', index=False)
```
