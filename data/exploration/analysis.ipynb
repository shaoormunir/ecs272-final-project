```python
import pandas as pd
# import geopandas as gpd
# from shapely.geometry import Point
```

```python
df_geo = pd.read_csv('geography.csv')

df_vac = pd.read_csv('vaccinations.csv')

df_ep = pd.read_csv('epidemiology.csv')

df_index = pd.read_csv('index.csv')

df_econ = pd.read_csv('economy.csv')

df_health = pd.read_csv('healh.csv')

df_demo = pd.read_csv('demographics.csv')
```

```python
df_ep
```

```python
df_ep = df_ep [['location_key', 'new_confirmed', 'new_deceased', 'new_recovered', 'new_tested']]
```

```python
df_ep = df_ep.groupby(['location_key']).agg({'new_confirmed': 'sum', 'new_deceased': 'sum', 'new_recovered': 'sum', 'new_tested': 'sum'}).reset_index()
```

```python
df_vac.head()
```

```python
df_vac = df_vac[['location_key', 'new_persons_fully_vaccinated']]
```

```python
df_vac = df_vac.groupby('location_key').sum().reset_index()
```

```python
df_vac
```

```python
df_index.head()
```

```python
df_index = df_index[
    ["location_key", "country_name", "subregion1_name", "subregion2_name", "iso_3166_1_alpha_3"]
]
```

```python
df_health.head()
```

```python
df_health = df_health[['location_key', 'smoking_prevalence', 'diabetes_prevalence', 'hospital_beds_per_1000', 'nurses_per_1000', 'physicians_per_1000', 'health_expenditure_usd']]
```

```python
df_econ.head()
```

```python
df_econ = df_econ[['location_key', 'gdp_per_capita_usd', 'human_capital_index']]
```

```python
df_demo.head()
```

```python
df_demo = df_demo[['location_key', 'population', 'population_density', 'human_development_index']]
```

```python
df_combined = pd.merge(df_index, df_econ, on='location_key')

df_combined = pd.merge(df_combined, df_ep, on='location_key')

df_combined = pd.merge(df_combined, df_health, on='location_key')

df_combined = pd.merge(df_combined, df_demo, on='location_key')

df_combined = pd.merge(df_combined, df_vac, on=['location_key'], how='left')
```

```python
df_combined
```

```python
# find new_confirmed, new_deceased, new_recovered, new_tested, new_persons_fully_vaccinated per 1000 people
df_combined['confirmed_per_1000'] = df_combined['new_confirmed'] / df_combined['population'] * 1000

df_combined['deceased_per_1000'] = df_combined['new_deceased'] / df_combined['population'] * 1000

df_combined['recovered_per_1000'] = df_combined['new_recovered'] / df_combined['population'] * 1000

df_combined['tested_per_1000'] = df_combined['new_tested'] / df_combined['population'] * 1000

df_combined['fully_vaccinated_per_1000'] = df_combined['new_persons_fully_vaccinated'] / df_combined['population'] * 1000
```

```python
df_combined
```

```python
df_combined.to_csv('exploration.csv', index=False)
```
