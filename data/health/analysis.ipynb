```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df_vac = pd.read_csv('vaccinations.csv')

df_ep = pd.read_csv('epidemiology.csv')

df_index = pd.read_csv('index.csv')

df_health = pd.read_csv('health.csv')

df_demo = pd.read_csv('demographics.csv')
```

```python
df_ep
```

```python
df_ep = df_ep[['location_key', 'date', 'new_confirmed', 'new_deceased', 'new_recovered', 'new_tested']]
```

```python
df_ep = df_ep.groupby(['location_key']).agg(
        total_deaths=("new_deceased", "sum"),
        total_recoveries=("new_recovered", "sum"),
        total_tested=("new_tested", "sum"),
        total_active_cases=("new_confirmed", "sum"),
    ).reset_index()
```

```python
df_ep
```

```python
df_vac.head()
```

```python
df_vac = df_vac.groupby(["location_key"]).agg(
        total_vaccinated=("new_persons_fully_vaccinated", "sum"),
).reset_index()
```

```python
df_vac
```

```python
df_index.head()
```

```python
df_index = df_index[
    ["location_key", "country_name", "iso_3166_1_alpha_2"]
]
df_index
```

```python
df_health["location_key"].value_counts()

# The dataframe does not contain "date" column
```

```python
df_health = df_health[['location_key', 'smoking_prevalence', 'diabetes_prevalence', 'hospital_beds_per_1000', 'nurses_per_1000', 'physicians_per_1000', 'health_expenditure_usd']]

df_health = df_health.groupby(["location_key"]).agg(
        mean_smoking_prevalence=("smoking_prevalence", "mean"),
        mean_diabetes_prevalence=("diabetes_prevalence", "mean"),
        mean_hospital_beds_per_1000=("hospital_beds_per_1000", "mean"),
        mean_nurses_per_1000=("nurses_per_1000", "mean"),
        mean_physicians_per_1000=("physicians_per_1000", "mean"),
        mean_health_expenditure_usd=("health_expenditure_usd", "mean"),
    ).reset_index()
```

```python
df_demo.head()

# This as the dataframe does not contain "date" column
```

```python
df_demo = df_demo[['location_key', 'population', 'population_density', 'human_development_index']]

df_demo = df_demo.groupby(["location_key"]).agg(
        population=("population", "sum"),
        population_density=("population_density", "mean"),
        human_development_index=("human_development_index", "mean"),
    ).reset_index()
```

```python
df_combined = pd.merge(df_index, df_ep, on="location_key")

df_combined = pd.merge(df_combined, df_health, on="location_key")

df_combined = pd.merge(df_combined, df_demo, on="location_key")

df_combined = pd.merge(df_combined, df_vac, on=["location_key"], how="left")

df_combined.drop(
    df_combined[df_combined["location_key"] != df_combined["iso_3166_1_alpha_2"]].index,
    inplace=True,
)

drop_cols = ["iso_3166_1_alpha_2", "location_key"]
df_combined.drop(drop_cols, axis=1, inplace=True)
```

```python
print(len(df_combined.columns))
print(df_combined.columns)
df_combined.fillna(0, inplace=True)
df_combined
```

```python
df_combined['total_cases_per_1000'] = df_combined['total_active_cases'] / df_combined['population'] * 1000

df_combined['total_deaths_per_1000'] = df_combined['total_deaths'] / df_combined['population'] * 1000

df_combined['total_recoveries_per_1000'] = df_combined['total_recoveries'] / df_combined['population'] * 1000

df_combined['total_tested_per_1000'] = df_combined['total_tested'] / df_combined['population'] * 1000
```

```python
df_combined.fillna(0, inplace=True)
```

```python
df_combined.to_csv('health-data.csv', index=False)
```
