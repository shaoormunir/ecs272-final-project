```python
import pandas as pd
```

```python
df_vac = pd.read_csv('vaccinations.csv')

df_ep = pd.read_csv('epidemiology.csv')

df_index = pd.read_csv('index.csv')

df_government = pd.read_csv('government-responses.csv')
```

```python
df_government.loc[df_government['location_key'] == 'US',:]
df_government.head()
```

```python
# drop mobility_retail_and_recreation column and convert date to datetime
df_govt = df_government.drop(columns=['fiscal_measures','international_support','income_support','emergency_investment_in_healthcare','investment_in_vaccines','stringency_index'])
df_govt.head()
```

```python
df_vac.head()
```

```python
df_vac = df_vac[['location_key', 'date', 'new_persons_fully_vaccinated']]
```

```python
df_ep.head()
```

```python
df_ep = df_ep[['location_key', 'date', 'new_confirmed', 'new_deceased', 'new_recovered']]
```

```python
df_index.columns
```

```python
df_index = df_index[
    ["location_key", "iso_3166_1_alpha_2", "country_name"]
]
df_index.value_counts()
```

```python
df_combined = pd.merge(df_index, df_vac, on=["location_key"], how="left")

df_combined = pd.merge(df_combined, df_ep, on=["location_key", "date"], how="left")

df_combined = pd.merge(df_combined, df_govt, on=["location_key", "date"], how="left")

df_combined.drop(df_combined[df_combined['location_key'] != df_combined['iso_3166_1_alpha_2']].index, inplace=True)
```

```python
df_combined
```

```python
df_combined = df_combined[['date', 'country_name', 'new_persons_fully_vaccinated',
       'new_confirmed', 'new_deceased', 'new_recovered', 'school_closing',
       'workplace_closing', 'cancel_public_events',
       'restrictions_on_gatherings', 'public_transport_closing',
       'stay_at_home_requirements', 'restrictions_on_internal_movement',
       'international_travel_controls', 'debt_relief',
       'public_information_campaigns', 'testing_policy', 'contact_tracing',
       'facial_coverings', 'vaccination_policy']]
df_combined = df_combined[~df_combined.date.isna()]
df_combined = df_combined[~df_combined.country_name.isna()]
df_combined.fillna(0, inplace=True)
```

```python
df_combined['date'] = pd.to_datetime(df_combined['date'])
df_combined = df_combined.sort_values(by=['country_name', 'date'])

# Task 2: Create additional columns for start and end
columns_to_process = ['school_closing', 'workplace_closing', 'cancel_public_events',
                       'restrictions_on_gatherings', 'public_transport_closing',
                       'stay_at_home_requirements', 'restrictions_on_internal_movement',
                       'international_travel_controls', 'debt_relief',
                       'public_information_campaigns', 'testing_policy', 'contact_tracing',
                       'facial_coverings', 'vaccination_policy']

for column in columns_to_process:
    df_combined[column].replace("0", "NoRestrictions", inplace=True)
    df_combined[column].replace("1", "NoRestrictions", inplace=True)
    df_combined[column].replace("2", "Restrictions", inplace=True)
    df_combined[column].replace("3", "Restrictions", inplace=True)
    df_combined[column].replace("4", "Restrictions", inplace=True)
    df_combined[column].replace("5", "Restrictions", inplace=True)

    df_combined[column].replace(0.0, "NoRestrictions", inplace=True)
    df_combined[column].replace(1.0, "NoRestrictions", inplace=True)
    df_combined[column].replace(2.0, "Restrictions", inplace=True)
    df_combined[column].replace(3.0, "Restrictions", inplace=True)
    df_combined[column].replace(4.0, "Restrictions", inplace=True)
    df_combined[column].replace(5.0, "Restrictions", inplace=True)

df_combined
```
